```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import *

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
from sklearn.datasets import load_iris

dataset = load_iris()

features = dataset.feature_names
train = pd.DataFrame(dataset.data, columns=features)
target = dataset.target

train
```

```python
data_scaler = MinMaxScaler(feature_range=(0,1))

train_fit = pd.DataFrame(data_scaler.fit_transform(train), columns=features)

train_fit
```

```python
X = train_fit
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
dtc_pred = y_pred

result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
clf = RandomForestClassifier(n_estimators=100)
clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
rfc_pred = y_pred


print("Features Importance:")
for i in range(len(features)):
    print(str(features[i]) + ' : ' + str(clf.feature_importances_[i]))

result = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(result)

result1 = classification_report(y_test, y_pred)
print("\nClassification Report:",)
print (result1)

result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
clf = KNeighborsClassifier()
clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
knc_pred = y_pred


result = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(result)

result1 = classification_report(y_test, y_pred)
print("\nClassification Report:",)
print (result1)

result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
# https://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html

avg_pred = []
for i in range(len(dtc_pred)):
    tmp = [dtc_pred[i], rfc_pred[i], knc_pred[i]]
    
    if(tmp[0] != tmp[1] or tmp[1] != tmp[2]):
        print(tmp)
    
    mode = max(set(tmp), key=tmp.count)
    
    avg_pred.append(mode if mode else tmp[0])


result = accuracy_score(y_test,avg_pred)
print("Accuracy:",result)
```
