```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import *

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
# https://scikit-learn.org/stable/auto_examples/ensemble/plot_voting_regressor.html

from sklearn.datasets import fetch_california_housing

dataset = fetch_california_housing(data_home='data')

features = dataset.feature_names
train = pd.DataFrame(dataset.data, columns=features)
target = dataset.target

print(dataset.DESCR)
```

```python
train
```

```python
data_scaler = MinMaxScaler(feature_range=(0,1))

train_fit = pd.DataFrame(data_scaler.fit_transform(train), columns=features)

train_fit
```

```python
X = train_fit
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
reg = RandomForestRegressor(random_state=1, n_estimators=50)
reg.fit(X_train, y_train)

y_pred = reg.score(X_test, y_test)
y_pred
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)

y_pred = reg.score(X_test, y_test)
y_pred
```

```python
reg = GradientBoostingRegressor(random_state=1, n_estimators=100)
reg.fit(X_train, y_train)

y_pred = reg.score(X_test, y_test)
y_pred
```
